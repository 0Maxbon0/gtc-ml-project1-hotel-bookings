```python
def detect_outliers_iqr(dataf, column_name):
    data = dataf[column_name]
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = dataf[(data < lower_bound) | (data > upper_bound)]

    print(f"\n IQR Outlier Report for {column_name}")
    print(f"Lower Bound: {lower_bound:}")
    print(f"Upper Bound: {upper_bound:}")
    print(f"Number of outliers detected: {len(outliers)}")
    print("-" * 30)

    return outliers
lead_time = detect_outliers_iqr(df, 'lead_time')
adr = detect_outliers_iqr(df, 'adr')
```

```python
#so company has 112593 null record or 94.31% of the records is null , lets fill it up with 0 becuase it contains a float numbers
df["company"].dtype # returns float
df["company"] = df["company"].fillna(0)
df["company"].isnull().sum()
```

```python
#agent has 16340 null values or 13.69% , now filling up agent with 0 becuase its type is float
df["agent"].dtype #returns float
df["agent"] = df["agent"].fillna(0)
df["agent"].isnull().sum()
```

```python
#country has 488 or 0.41% null values , using mode will fill up the missing records
df["country"].dtype
df["country"] = df["country"].mode()[0]
df["country"].isnull().sum()
```

```python
#children has only 4 or 0.003 null values , using meadian to fill up them
df["children"].sample()
df["children"] = df["children"].fillna(df["children"].median())
df["children"].isnull().sum()
```

```python
#checking the duplicated records
df.duplicated().sum()
df.duplicated().sum()/len(df)*100
```

```python
#we have a 33753 duplicated record or 28.27% record , droping them
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
#handling outliers
df["adr"].max()
df["adr"].mean()
```

```python
#capping adr and lead time according to the IQR

# IQR Outlier Report for lead_time
# Lower Bound: -195.0
# Upper Bound: 373.0
# Number of outliers detected: 3005
# ------------------------------

#  IQR Outlier Report for adr
# Lower Bound: -15.774999999999991
# Upper Bound: 211.065
# Number of outliers detected: 3793
df.loc[df['adr'] > 211.1, 'adr'] = 211.1
df.loc[df['adr'] < 0, 'adr'] = 0
df.loc[df['lead_time'] > 373, 'lead_time'] = 373
df.loc[df['lead_time'] < -195, 'lead_time'] = -195
#i used IQR outliers boundaries to cap the records becuese it shows me the boundaries as numbers so i can limit the data in more realistic zone
# Why making the adr minimum a zero ? becuase the adr should be a positive number by defualt --->(Total Room Revenue) / (Number of Occupied Rooms). thats cant generate a negative number
```
